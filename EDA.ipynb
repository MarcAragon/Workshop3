```python
fig = plt.figure(figsize=(12, 8))
fig.suptitle("Target variable distribution across years", fontsize=16)

#2x6 grid
gs = GridSpec(2, 6, figure=fig)

#Top row
ax1 = fig.add_subplot(gs[0, 0:2])
ax2 = fig.add_subplot(gs[0, 2:4])
ax3 = fig.add_subplot(gs[0, 4:6])

#Bottom row
ax4 = fig.add_subplot(gs[1, 0:3])
ax5 = fig.add_subplot(gs[1, 3:6])

ax1.hist(Df2015['Happiness Score'], bins=20, color='black')

ax1.axvline(Df2015['Happiness Score'].mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {Df2015['Happiness Score'].mean():.2f}')
ax1.axvline(Df2015['Happiness Score'].median(), color='blue', linestyle='-', linewidth=2, label=f'Median: {Df2015['Happiness Score'].median():.2f}')

ax1.set_title("2015")
ax1.set_xlabel("Happiness Score")
ax1.set_ylabel("Frequency")
ax1.legend()

ax2.hist(Df2016['Happiness Score'], bins=20, color='black')

ax2.axvline(Df2016['Happiness Score'].mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {Df2016['Happiness Score'].mean():.2f}')
ax2.axvline(Df2016['Happiness Score'].median(), color='blue', linestyle='-', linewidth=2, label=f'Median: {Df2016['Happiness Score'].median():.2f}')

ax2.set_title('2016')
ax2.set_xlabel('Happiness Score')
ax2.set_ylabel('Frecuency')
ax2.legend()

ax3.hist(Df2017['Happiness.Score'], bins=20, color='black')

ax3.axvline(Df2017['Happiness.Score'].mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {Df2017['Happiness.Score'].mean():.2f}')
ax3.axvline(Df2017['Happiness.Score'].median(), color='blue', linestyle='-', linewidth=2, label=f'Median: {Df2017['Happiness.Score'].median():.2f}')

ax3.set_title('2017')
ax3.set_xlabel('Happiness Score')
ax3.set_ylabel('Frecuency')
ax3.legend()

ax4.hist(Df2018['Score'], bins=20, color='black')

ax4.axvline(Df2018['Score'].mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {Df2018['Score'].mean():.2f}')
ax4.axvline(Df2018['Score'].median(), color='blue', linestyle='-', linewidth=2, label=f'Median: {Df2018['Score'].median():.2f}')

ax4.set_title('2018')
ax4.set_xlabel('Happiness Score')
ax4.set_ylabel('Frecuency')
ax4.legend()

ax5.hist(Df2019['Score'], bins=20, color='black')

ax5.axvline(Df2019['Score'].mean(), color='red', linestyle='--', linewidth=2, label=f'Mean: {Df2019['Score'].mean():.2f}')
ax5.axvline(Df2019['Score'].median(), color='blue', linestyle='-', linewidth=2, label=f'Median: {Df2019['Score'].median():.2f}')

ax5.set_title('2019')
ax5.set_xlabel('Happiness Score')
ax5.set_ylabel('Frecuency')
ax5.legend()

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(14, 8))
fig.suptitle("Higher valuation for GDP per capita across years", fontsize=16)

#2x6 grid
gs = GridSpec(2, 6, figure=fig)

#Top row
ax1 = fig.add_subplot(gs[0, 0:2])
ax2 = fig.add_subplot(gs[0, 2:4])
ax3 = fig.add_subplot(gs[0, 4:6])

#Bottom row
ax4 = fig.add_subplot(gs[1, 0:3])
ax5 = fig.add_subplot(gs[1, 3:6])

#Top 10 countries with the higher valuation for GDP per capita
BiggestGDP10 = Df2015.nlargest(10, 'Economy (GDP per Capita)')
BiggestGDP10.sort_values('Economy (GDP per Capita)', ascending=True, inplace=True)
ax1.barh(BiggestGDP10['Country'], BiggestGDP10['Economy (GDP per Capita)'], color='black')
ax1.set_title('2015')
ax1.set_xlabel('Percentage of the total happiness score')

BiggestGDP10 = Df2016.nlargest(10, 'Economy (GDP per Capita)')
BiggestGDP10.sort_values('Economy (GDP per Capita)', ascending=True, inplace=True)
ax2.barh(BiggestGDP10['Country'], BiggestGDP10['Economy (GDP per Capita)'], color='black')
ax2.set_title('2016')
ax2.set_xlabel('Percentage of the total happiness score')

BiggestGDP10 = Df2017.nlargest(10, 'Economy..GDP.per.Capita.')
BiggestGDP10.sort_values('Economy..GDP.per.Capita.', ascending=True, inplace=True)
ax3.barh(BiggestGDP10['Country'], BiggestGDP10['Economy..GDP.per.Capita.'], color='black')
ax3.set_title('2017')
ax3.set_xlabel('Percentage of the total happiness score')

BiggestGDP10 = Df2018.nlargest(10, 'GDP per capita')
BiggestGDP10.sort_values('GDP per capita', ascending=True, inplace=True)
ax4.barh(BiggestGDP10['Country or region'], BiggestGDP10['GDP per capita'], color='black')
ax4.set_title('2018')
ax4.set_xlabel('Percentage of the total happiness score')

BiggestGDP10 = Df2019.nlargest(10, 'GDP per capita')
BiggestGDP10.sort_values('GDP per capita', ascending=True, inplace=True)
ax5.barh(BiggestGDP10['Country or region'], BiggestGDP10['GDP per capita'], color='black')
ax5.set_title('2019')
ax5.set_xlabel('Percentage of the total happiness score')

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(12, 8))
fig.suptitle("Economy vs family comparison across years", fontsize=16)

#2x6 grid
gs = GridSpec(2, 6, figure=fig)

#Top row
ax1 = fig.add_subplot(gs[0, 0:2])
ax2 = fig.add_subplot(gs[0, 2:4])
ax3 = fig.add_subplot(gs[0, 4:6])

#Bottom row
ax4 = fig.add_subplot(gs[1, 0:3])
ax5 = fig.add_subplot(gs[1, 3:6])

#Economy vs Family comparison
Boxplot = ax1.boxplot([Df2015['Economy (GDP per Capita)'], Df2015['Family']], tick_labels=['Economy (GDP per Capita)', 'Family'], patch_artist=True)
ax1.set_title('2015')
ax1.set_ylabel('% of the total happiness score')
Colors = ['black', 'gray']
for Patch, Color in zip(Boxplot['boxes'], Colors):
    Patch.set_facecolor(Color)
ax1.grid(True, axis='y', linestyle='--', alpha=0.7)


Boxplot = ax2.boxplot([Df2016['Economy (GDP per Capita)'], Df2016['Family']], tick_labels=['Economy (GDP per Capita)', 'Family'], patch_artist=True)
ax2.set_title('2016')
ax2.set_ylabel('% of the total happiness score')
Colors = ['black', 'gray']
for Patch, Color in zip(Boxplot['boxes'], Colors):
    Patch.set_facecolor(Color)
ax2.grid(True, axis='y', linestyle='--', alpha=0.7)

Boxplot = ax3.boxplot([Df2017['Economy..GDP.per.Capita.'], Df2017['Family']], tick_labels=['Economy (GDP per Capita)', 'Family'], patch_artist=True)
ax3.set_title('2017')
ax3.set_ylabel('% of the total happiness score')
Colors = ['black', 'gray']
for Patch, Color in zip(Boxplot['boxes'], Colors):
    Patch.set_facecolor(Color)
ax3.grid(True, axis='y', linestyle='--', alpha=0.7)

Boxplot = ax4.boxplot([Df2018['GDP per capita'], Df2018['Social support']], tick_labels=['Economy (GDP per Capita)', 'Family'], patch_artist=True)
ax4.set_title('2018')
ax4.set_ylabel('% of the total happiness score')
Colors = ['black', 'gray']
for Patch, Color in zip(Boxplot['boxes'], Colors):
    Patch.set_facecolor(Color)
ax4.grid(True, axis='y', linestyle='--', alpha=0.7)

Boxplot = ax5.boxplot([Df2019['GDP per capita'], Df2019['Social support']], tick_labels=['Economy (GDP per Capita)', 'Family'], patch_artist=True)
ax5.set_title('2019')
ax5.set_ylabel('% of the total happiness score')
Colors = ['black', 'gray']
for Patch, Color in zip(Boxplot['boxes'], Colors):
    Patch.set_facecolor(Color)
ax5.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
#Regiones con el puntaje mas alto en promedio en este periodo
RegionMeansDf = Df.groupby('Region').agg({
    'Happiness_Score': 'mean',
    'Economy_GDP_per_Capita': 'mean',
    'Family': 'mean'
})

plt.barh(RegionMeansDf.sort_values('Happiness_Score').index, RegionMeansDf.sort_values('Happiness_Score')['Happiness_Score'].values, color='black')
plt.title('Average happiness score across all years by region')
plt.xlabel('Avg')
plt.show()
```

```python
plt.barh(RegionMeansDf.sort_values('Economy_GDP_per_Capita').index, RegionMeansDf.sort_values('Economy_GDP_per_Capita')['Economy_GDP_per_Capita'].values, color='black')
plt.title('Average valuation of the country economy across all years by region')
plt.xlabel('Avg total of the happiness score')
plt.show()
```

```python
plt.barh(RegionMeansDf.sort_values('Family').index, RegionMeansDf.sort_values('Family')['Family'].values, color='black')
plt.title('Average valuation of the social aspect across all years by region')
plt.xlabel('Avg total of the happiness score')
plt.show()
```
